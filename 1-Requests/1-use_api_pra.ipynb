```python
# 引入 requests 模組
import requests

# 測試用 API 網址
url = 'https://httpbin.org/'

# 發送 GET 請求並取得回應
response = requests.get(url)
if response.status_code != 200:
    print("回傳異常")

# 回應的狀態碼
print(response.status_code)

# 回應的原始內容（位元組）
print(response.content)

# 回應的文字內容
print(response.text)
```

```python
# 使用 POST 搭配 data (常用在表單提交)
r = requests.post(url+'post', data={"username": "Johnwalker"})
print(r.json())
```

```python
# 使用 POST 搭配 JSON (常用在 API 資料傳輸)
r = requests.post(url+'post', json={"username": "Johnwalker"})
print(r.json())
```
