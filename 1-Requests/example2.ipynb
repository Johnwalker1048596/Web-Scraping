```python
import json
import os
import requests
from bs4 import BeautifulSoup as bs

folder_name = "line_stickers"

# 建立資料夾
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print(f"已建立資料夾 {folder_name}")
else:
    print(f"資料夾已存在 {folder_name}")

# LINE 貼圖網址
url = "https://store.line.me/stickershop/product/35807/zh-Hant"

r = requests.get(url)
soup = bs(r.text, "lxml")

# 找到貼圖清單 <ul>
sticker_list = soup.find("ul", class_="mdCMN09Ul FnStickerList")

# 找到所有貼圖 <li>
sticker_elements = sticker_list.find_all('li')
print(f"總共發現 {len(sticker_elements)} 張貼圖\n")

# 逐張下載
for index, sticker_element in enumerate(sticker_elements, start=1):

    # 從 data-preview 解析出圖片 URL
    data_preview = json.loads(sticker_element["data-preview"])
    static_url = data_preview["staticUrl"]
    print(f"[{index}] 下載貼圖：{static_url}")

    # 下載圖片
    image_content = requests.get(static_url).content

    # 儲存圖片
    file_path = f"{folder_name}/line_sticker_{index}.png"
    with open(file_path, "wb") as f:
        f.write(image_content)

    print(f"已下載：{file_path}")

print("\n全部下載完成！")
```

```python
import json
import os
import requests
from bs4 import BeautifulSoup as bs

folder_name = "line_stickers"

# 建立主資料夾
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print(f"已建立 {folder_name}")
else:
    print(f"資料夾已存在 {folder_name}")


def get_stickers(url):

    # 下載網頁
    r = requests.get(url)
    soup = bs(r.text, "lxml")

    # 取得貼圖包名稱
    sticker_name = soup.find("p", class_="mdCMN38Item01Ttl").text

    # 建立該貼圖包的資料夾
    pack_path = f"{folder_name}/{sticker_name}"
    if not os.path.exists(pack_path):
        os.mkdir(pack_path)
        print(f"已建立資料夾：{sticker_name}")
    else:
        print(f"資料夾已存在：{sticker_name}")

    # 找到貼圖列表
    sticker_list = soup.find("ul", class_="mdCMN09Ul FnStickerList")

    # 找到每張貼圖 li
    sticker_elements = sticker_list.find_all("li")
    print(f"共找到 {len(sticker_elements)} 張貼圖\n")

    # 逐張下載
    for index, sticker_element in enumerate(sticker_elements, start=1):

        # data-preview 裡是 JSON，包含 staticUrl
        data_preview = json.loads(sticker_element["data-preview"])
        static_url = data_preview["staticUrl"]

        print(f"[{index}] 下載中：{static_url}")

        # 下載貼圖
        image_content = requests.get(static_url).content

        # 儲存圖片
        file_path = f"{pack_path}/line_sticker_{index}.png"
        with open(file_path, "wb") as f:
            f.write(image_content)

        print(f"已下載：{file_path}")

    print("\n下載完成！")


# ▶ 你要下載的貼圖網址（直接放下面這行）
get_stickers("https://store.line.me/stickershop/product/28419/zh-Hant")
```
